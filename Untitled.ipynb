# groupBy와 pivot_table로 다양한 집계 연산 하기

## groupby

```python
# 2개 이상의 데이터를 보기 위해서 groupby와 pivot_table을 사용합니다.
df.groupby?
#새로운 기능을 사용해볼 때는 ?를 써서 example을 확인하는 것을 권장드립니다.
```

```python
df.groupby(['성별코드']).mean() 
#성별코드 값에 따라 모든 열에 대한 평균을 구합니다.남성, 여성에 따른 데이터를 세어서 보여줍니다.
```

```python
df.groupby(['성별코드']).count() #성별코드 값에 따라 데이터를 세어보기
```

```python
df.groupby(['성별코드'])['가입자일련번호'].count()
#결과는 같음 df['성별코드'].value_counts()
```

```python
df.groupby(['성별코드','음주여부'])['가입자일련번호'].count()
```

```python
#성별코드와 음주여부로 그룹화 하고  rGTP의 평균을 구함.
df.groupby(['성별코드','음주여부'])['감마지티피']. mean()
```

```python
#성별코드와 음주여부로 그룹화 하고  rGTP의 요약수치를 구함.
df.groupby(["성별코드", "음주여부"])["감마지티피"].describe()
```

```python
#agg을 사용하면 여러수치를 함께 구할 수 있습니다.
df.groupby(["성별코드", "음주여부"])["감마지티피"].agg(["count", "mean", "median"])
```

## pivot table

```python
'''pivot_table에서 할 수 있는 것은 groupby에서 대부분 할 수 있고, 
groupby에서 할 수 있는 것은 pivot_table에서 할 수 있습니다.
pivot_table의 연산 속도가 groupby보다 빠르고, 사용 방법이 직관적입니다.'''
df.pivot #pivot은 연산을 하지 않고 데이터의 구조를 바꾸고자 할 때 사용합니다.
```

```python
df.pivot_table(index='음주여부')
```

```python
df.pivot_table(index="성별코드")
```

```python
df.pivot_table(index="성별코드", values="가입자일련번호") #기본은 평균값이 나옴
```

```python
df.pivot_table(index="성별코드", values="가입자일련번호", aggfunc="count")
```

```python
pd.pivot_table(df, index="음주여부", values="감마지티피")
```

```python
pd.pivot_table(df, index="음주여부", values="감마지티피", aggfunc=["mean", "median"])
```

```python
pd.pivot_table(df, index="음주여부", values="감마지티피", aggfunc=["describe"])
```

```python
pd.pivot_table(df, index=["음주여부",'성별코드'], values="감마지티피", aggfunc=["describe"])
```
